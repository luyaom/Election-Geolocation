```python
import pandas as pd
```

### 1. Read original data

```python
election_data = pd.read_csv('US_County_Level_Election_Results_08-16-master/US_County_Level_Presidential_Results_08-16.csv', dtype={'fips_code':str})
print(election_data.shape)
print(election_data.head())
```

### 2. Calculate election ratio 

```python
election_data['2008_dem_ratio'] = election_data['dem_2008'] / election_data['total_2008']
election_data['2012_dem_ratio'] = election_data['dem_2012'] / election_data['total_2012']
election_data['2016_dem_ratio'] = election_data['dem_2016'] / election_data['total_2016']
```

```python
election_data['2008_gop_ratio'] = election_data['gop_2008'] / election_data['total_2008']
election_data['2012_gop_ratio'] = election_data['gop_2012'] / election_data['total_2012']
election_data['2016_gop_ratio'] = election_data['gop_2016'] / election_data['total_2016']
```

```python
election_data['2008_oth_ratio'] = election_data['oth_2008'] / election_data['total_2008']
election_data['2012_oth_ratio'] = election_data['oth_2012'] / election_data['total_2012']
election_data['2016_oth_ratio'] = election_data['oth_2016'] / election_data['total_2016']
```

### 3. Calculate election result

```python
election_data['2008_result'] = ['dem' if dem > gop else 'gop' for (dem, gop) in zip(election_data['dem_2008'], election_data['gop_2008'])]
election_data['2012_result'] = ['dem' if dem > gop else 'gop' for (dem, gop) in zip(election_data['dem_2012'], election_data['gop_2012'])]
election_data['2016_result'] = ['dem' if dem > gop else 'gop' for (dem, gop) in zip(election_data['dem_2016'], election_data['gop_2016'])]
```

```python
election_data.head() ## finalized data
```

```python
election_data.to_csv('election_data.csv', index=False)
```
